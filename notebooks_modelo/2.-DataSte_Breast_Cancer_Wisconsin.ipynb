## 1.- Carga del DataSet

```python
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()

X = data.data
y = data.target

print("Shape: ", X.shape)
```

## 2.- Escalamiento de los datos

En este caso al estar usando **KNN** es obligatorio estar realizando un escalamiento 


```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

## 3.- Entrenamiento de ambos modelos


```python
from sklearn.neighbors import KNeighborsClassifier

knn_euclidean = KNeighborsClassifier(n_neighbors=11, metric="euclidean")
knn_manhattan = KNeighborsClassifier(n_neighbors=11, metric="manhattan")

knn_euclidean.fit(X_train, y_train)
knn_manhattan.fit(X_train,y_train)

print("Euclidean: ", knn_euclidean.score(X_test, y_test))
print("Manhattan: ", knn_manhattan.score(X_test, y_test))
```

# VALIDACÍON CRUZADA

```python
from sklearn.model_selection import cross_val_score
```

```python
knn_euclidean = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn_manhattan = KNeighborsClassifier(n_neighbors=5, metric='manhattan')

scores_euclidean = cross_val_score(knn_euclidean, X, y, cv=10)
scores_manhattan = cross_val_score(knn_manhattan, X, y, cv=10)

print("Euclidean promedio:", scores_euclidean.mean())
print("Manhattan promedio:", scores_manhattan.mean())
```
